```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("zomato_cleaned_2000.csv")
print(dataframe)
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in_type'])
plt.xlabel('type of resturant')
```

Conclusion: The majority of the restaurants fall into the dining category

Dining restaurants are preferred by a larger number of individuals

```python
grouped_data = dataframe.groupby('listed_in_type') ['votes'].sum()

result = pd.DataFrame({'votes': grouped_data})

plt.plot(result, c="green", marker="o")

plt.xlabel("Type of restaurant", c="red", size=20)

plt.ylabel("Votes", c="red", size=20)
```

```python
plt.hist(dataframe['rate'], bins=5)

plt.title("Ratings Distribution")

plt.show()
```

Conclusion: The majority of restaurants received ratings ranging from 3.5 to 4

the majority of couples prefer restaurants withi an approximate cost of 300 rupees

```python
couple_data = dataframe['approx_costfor_two_people'] 
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize =(6,6))

sns.boxplot(x='online_order', y ='rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in_type', columns='online_order',aggfunc='size', fill_value=0) 

sns.heatmap(pivot_table, annot= True, cmap ="YlGnBu", fmt='d')

plt.title("Heatmap")

plt.xlabel("Online Onder")

plt.ylabel("Listed In Type")

plt.show()
```

Conclusion: Dining restaurants primarily accept offline orders, whereas cafes primarily
accept offline orders, whereas cafes primarily receive online orders. This suggests that 
clients prefer to place orders in person at restaurants, but prefer online ordering 
at cafes.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='approx_costfor_two_people', y='rate', data=dataframe, alpha=0.6)
plt.title('Relationship Between Cost for Two and Rating')
plt.xlabel('Average Cost for Two ()')
plt.ylabel('Customer Rating')
plt.show()
```

Conclusion: 

the scatter plot shows no strong linear relationship between the cost for two and customer rating.

Restaurants with both low and high costs can achieve high ratings, indicating that price alone doesnt decide quality.

However, a slight concentration of higher ratings can be observed around moderate-cost restaurants, suggesting that customers might find better value for money in that range.

Overall, service quality, taste, and experience seem to influence ratings more than price.
